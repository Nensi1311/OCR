```python
import fitz  # PyMuPDF
from PIL import Image

def pdf_to_image(pdf_path, page_num=0):
    """Converts a specific PDF page to a PIL Image object."""
    try:
        doc = fitz.open(pdf_path)
        page = doc.load_page(page_num)

        # Set a high resolution (e.g., 300 DPI) for better OCR accuracy
        zoom = 3
        mat = fitz.Matrix(zoom, zoom)
        pix = page.get_pixmap(matrix=mat, alpha=False)

        # Convert to a PIL Image
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        return img
    except Exception as e:
        print(f"Error converting PDF to image: {e}")
        return None

# Example usage (assuming 'statement.pdf' is in the same directory)
image = pdf_to_image('SENDHA RAM HDFC-1751718779796.pdf', page_num=0)
if image:
    image.save("page_1.png") # Optional: save to check the output
    print("PDF page converted to image.")
```

```python
import easyocr
import numpy as np

def extract_text_with_ocr(image):
    """Uses EasyOCR to read text from the image."""
    # Initialize the OCR reader (choose languages, 'en' for English)
    reader = easyocr.Reader(['en'])

    np_img = np.array(image)

    results = reader.readtext(np_img)

    return results

# Example usage
if image:
    ocr_results = extract_text_with_ocr(image)
    all_text = ' '.join([res[1] for res in ocr_results])
    print(f"OCR Extracted Text Sample: {all_text[:6000]}...")
```

```python
import easyocr
from pdf2image import convert_from_path

def extract_text_from_pdf_easyocr(input_file, language_list=['en']):
    # Initialize the EasyOCR reader
    reader = easyocr.Reader(language_list)
    
    # Convert PDF pages to images
    pages = convert_from_path(input_file)
    
    for i, page in enumerate(pages, start=1):
        print(f"\n--- Page {i} ---\n")
        
        # Convert each page to a format suitable for OCR
        page.save("temp_page.jpg", "JPEG")
        
        # Perform OCR on the page image
        result = reader.readtext(f"page {i}.jpg", detail=0)
        
        if result:
            # Join all detected text segments
            text_content = "\n".join(result)
            print(text_content)
        else:
            print("[No text found on this page]")
```
